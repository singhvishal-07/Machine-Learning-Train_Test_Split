## Dividng the dataset into Training set and Testing set. This will help us to determine  the accuaracy of our dataset and help us when working with large data. 

#### For this we will be using the train_test_split of the model_selection

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn import linear_model
```

```python
data = pd.read_csv('TrainingSet.csv')
print(data.shape)
data
```

### Car MIleage VS Sell Price

```python
plt.rcParams['figure.figsize'] = (15.0,7.5)
plt.scatter(data['Mileage'],data['Sell Price($)'], label = 'Scatter Points')
plt.xlabel('Mileage')
plt.ylabel('Sell Price($)')
plt.legend()
```

### Car Age vs Sell Price

```python
plt.scatter(data['Age(yrs)'],data['Sell Price($)'], label = 'Scatter Points')
plt.xlabel('Age(yrs)')
plt.ylabel('Sell Price($)')
plt.legend()
```

```python
from mpl_toolkits import mplot3d
fig = plt.figure()
ax = plt.axes(projection="3d")
```

### Car Mileage VS Car Age VS Sell Price

```python
# Plotting the scatter points in all 3 Dimensions.
# Inorder to get the interactive 3D model -  enable the comment below!
# %matplotlib qt
fig = plt.figure()
ax = plt.axes(projection="3d")

ax.scatter3D(data['Mileage'],data['Age(yrs)'],data['Sell Price($)'], c='blue', label = 'Scatter Plots');
ax.legend()
ax.set_xlabel('Mileage')
ax.set_ylabel('Age(yrs)')
ax.set_zlabel('Sell Price($)')

plt.show()
```

### Thus we can observe that linear regression is possible for the data after viewing the scatter plot 
#### Thus we perform train_test_split

```python
x = data[['Mileage','Age(yrs)']]
x
```

```python
y = data['Sell Price($)']
y
```

```python
from sklearn.model_selection import train_test_split
model = linear_model.LinearRegression()
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.2)
```

```python
len(x_train)
```

```python
len(x_test)
```

#### We observe that the the given dataset is divided as 80% = training and 20% as testing
#### thus we have 16 values in Training_Set and 4 in Test_Set

### Also the values in the training set keeps changing and selected as random which is much better than keeping them constant 
### We can keep the data in training set as constant by adding attribute - 'random_state  = 10'

```python
x_train
```

```python
# Now performing the regression the given data

model.fit(x_train,y_train)
```

```python
model.predict(x_test)
```

```python
y_test
```

```python
model.score(x_test,y_test)
```

### Thus with the help of train_test_split we divided the data into 80% - training and  20% - testing sets and found the accuracy.
